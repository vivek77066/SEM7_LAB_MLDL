```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Salary_Data.csv")
X = data["YearsExperience"].values  # Shape (n,)
y = data["Salary"].values
```

```python
m = 0
c = 0
L = 0.01
epochs = 1000
n = len(X)

# Gradient Descent Loop
for i in range(epochs):
    y_pred = m * X + c
    D_m = (-2/n) * np.sum(X * (y - y_pred))
    D_c = (-2/n) * np.sum(y - y_pred)
    m = m - L * D_m
    c = c - L * D_c
```

```python
if i % 100 == 0:
        loss = np.mean((y - y_pred) ** 2)
        print(f"Epoch {i}: Loss={loss:.2f} m={m:.2f}, c={c:.2f}")
```

```python
plt.scatter(X, y, color='red', label='Data')
plt.plot(X, m * X + c, color='blue', label='Fitted Line')
plt.legend()
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression via Gradient Descent")
plt.show()

# Final line
print(f"Final equation: y = {m:.2f}x + {c:.2f}")
```
